# Ethnol prices: Paulinia BRL and USD
corr between BRL and USD ethanol based on Paulinia daily cotation from 2010


```python
dates= ['Data_EtOHhid_paulinia']
#dates.apply(lambda t: pd.to_datetime(t, format='%d/%m/%Y'))
df_rawdata_EtOH_paulinia = pd.read_excel(f'{pasta_raiz}/EDA_price_estimation.xlsx', sheet_name='rawdata_EtOH_paulinia', parse_dates=dates)
df_rawdata_EtOH_paulinia.head(5)
```

```python
df_rawdata_EtOH_paulinia = (
    df_rawdata_EtOH_paulinia
    .assign(Ano_EtOHhid_paulinia = df_rawdata_EtOH_paulinia['Data_EtOHhid_paulinia'].dt.year)
    .reset_index()
    )
```

```python
df_rawdata_EtOH_paulinia.info()
```

```python
fig = px.box(df_rawdata_EtOH_paulinia, y=['BRL_EtOHhid_paulinia', 'USD_EtOHhid_paulinia'], notched=False, template="simple_white",
            height=600, width=1200, color_discrete_sequence=px.colors.qualitative.Vivid, labels={'value':'BRL ou USD /m3'},
            title='Cotação para Etanol Hidratado - Paulinia SP (sem frete e impostos)')
fig.show()
```

```python
fig= px.scatter(df_rawdata_EtOH_paulinia, x="Data_EtOHhid_paulinia", y=['USD_EtOHhid_paulinia','BRL_EtOHhid_paulinia'],
                title='Cotação Diária EtOH hidratado, Paulinia SP - média móvel = 30 dias', height=600, width=1200,
                 trendline= "rolling" ,trendline_options=dict(window=30),trendline_scope='trace', trendline_color_override='black',
                 opacity=1, color_discrete_sequence=px.colors.qualitative.Vivid, template="simple_white",
                 labels={'variable':'cotação', 'value':'USD ou BRL/m3', 'Data_EtOHhid_paulinia':'dia/mês/ano'})

fig.update_layout(title_font_color="Black", font_size=(14),title_x=0.1)
fig.show()
```

```python
fig= px.scatter(df_rawdata_EtOH_paulinia, x="USD_EtOHhid_paulinia", y=['BRL_EtOHhid_paulinia'],
                title='Cotação EtOH hidratado, Paulinia SP: BRL vs. USD', height=600, width=1200,
                 trendline= "ols" , trendline_scope='trace', trendline_color_override='black', color='Ano_EtOHhid_paulinia',
                 opacity=1, color_discrete_sequence=px.colors.qualitative.Vivid, template="simple_white",
                 labels={'Ano_EtOHhid_paulinia':'Ano/cotação'})

fig.update_layout(title_font_color="Black", font_size=(14),title_x=0.1)
fig.show()
results = px.get_trendline_results(fig)
print(results)

results.px_fit_results.iloc[0].summary()
```

```python
df_rawdata_EtOH_paulinia1=df_rawdata_EtOH_paulinia.groupby(['Ano_EtOHhid_paulinia'])[['BRL_EtOHhid_paulinia']].describe().round(2)
#df_rawdata4=df_rawdata4.query('Ano_EtOHhidr_MT > 2012')
display(df_rawdata_EtOH_paulinia1)
```

# Corn Kernel Exploratory Data Analysis

# Corn Kernel Exploratory Data Analysis
*   Compare corn kernel cotation to Mato Grosso, Rondonia and Brazil




```python
variables=['Data_milho_MT', 'BRL_milho_MT', 'BRL_milho_BR', 'Data_milho_RO', 'BRL_milho_RO']
dates= ['Data_milho_MT', 'Data_milho_RO']
#dates.apply(lambda t: pd.to_datetime(t, format='%d/%m/%Y'))
df_rawdata_Corn = pd.read_excel(f'{pasta_raiz}/EDA_price_estimation.xlsx', sheet_name='rawdata_CornSoy', parse_dates=dates, usecols=variables)
df_rawdata_Corn.head(5)
```

```python
df_rawdata_Corn['BRL_milho_MT_ton']=df_rawdata_Corn['BRL_milho_MT']*16.667
df_rawdata_Corn['BRL_milho_BR_ton']=df_rawdata_Corn['BRL_milho_BR']*16.667
df_rawdata_Corn['BRL-milho_RO_ton']=df_rawdata_Corn['BRL_milho_RO']*16.667

df_rawdata_Corn = (
    df_rawdata_Corn
    .assign(Ano_milho = df_rawdata_Corn['Data_milho_MT'].dt.year)
    .reset_index()
    )
df_rawdata_Corn.info()
```

```python
df_rawdata_Corn1=(
    df_rawdata_Corn
    .groupby(['Ano_milho'])[['BRL_milho_MT', 'BRL_milho_BR']].describe().round(2)
    .query('Ano_milho > 2012')
)
display(df_rawdata_Corn1)
```

```python
df_rawdata_Corn2=(
    df_rawdata_Corn
    .groupby(['Ano_milho'])[['BRL_milho_RO', 'BRL_milho_BR']].describe().round(2)
    .query('Ano_milho > 2018')
)
display(df_rawdata_Corn2)
```

```python
fig = px.box(df_rawdata_Corn, y=['BRL_milho_MT', 'BRL_milho_RO', 'BRL_milho_BR' ], notched=False, template="simple_white",
            height=600, width=1200, color_discrete_sequence=px.colors.qualitative.Vivid, labels={'value':'BRL ou USD /60kg'},
            title='Cotação para Milho 60kg - MT; RO e BR')
fig.show()
```

```python
fig= px.scatter(df_rawdata_Corn, x="Data_milho_MT", y=['BRL_milho_MT','BRL_milho_RO', 'BRL_milho_BR'],
                title='Cotação Média Mensal: Milho 60kg - média móvel = 3 meses', height=600, width=1200,
                 trendline= "rolling" ,trendline_options=dict(window=3),trendline_scope='trace',
                opacity=1, color_discrete_sequence=px.colors.qualitative.Vivid, template="simple_white",
                 labels={'variable':'cotação', 'value':'BRL/60kg', 'Data_milho_MT':'dia/mês/ano'})
fig.update_layout(title_font_color="Black", font_size=(14),title_x=0.1)
fig.show()
```

# DDGS - Exploratory Data Analysis
Compare DDGS and soybean meal on protein bases

```python
variables=['Data_FSoja_MT', 'BRL_FSoja_MT', 'BRL_Fsoja_BR']
dates= ['Data_FSoja_MT']
#dates.apply(lambda t: pd.to_datetime(t, format='%d/%m/%Y'))
df_rawdata_FSoja = pd.read_excel(f'{pasta_raiz}/EDA_price_estimation.xlsx', sheet_name='rawdata_CornSoy', parse_dates=dates, usecols=variables)
df_rawdata_FSoja.head(5)
```

```python
#Soy protein = 45% and DDGS protein = 30%
df_rawdata_FSoja['BRL_DDGS_MT_estimado']=df_rawdata_FSoja['BRL_FSoja_MT'] * 0.667

df_rawdata_FSoja = (
    df_rawdata_FSoja
    .assign(Ano_FSoja = df_rawdata_FSoja['Data_FSoja_MT'].dt.year)
)

df_rawdata_FSoja1=(
    df_rawdata_FSoja
    .groupby(['Ano_FSoja'])[['BRL_FSoja_MT', 'BRL_DDGS_MT_estimado']].describe().round(2)
    .query('Ano_FSoja > 2012')
)
display(df_rawdata_FSoja1)
```

```python
fig = px.box(df_rawdata_FSoja, y=['BRL_DDGS_MT_estimado', 'BRL_FSoja_MT', 'BRL_Fsoja_BR'], notched=False, template="simple_white",
            height=600, width=1200, color_discrete_sequence=px.colors.qualitative.Vivid, labels={'value':'BRL/60kg', 'variable': 'cotações'},
            title='Cotação para Farelo de Soja e DDGS (estimado) - MT e BR')
fig.show()
```

```python
fig= px.scatter(df_rawdata_FSoja, x="Data_FSoja_MT", y=['BRL_FSoja_MT','BRL_Fsoja_BR', 'BRL_DDGS_MT_estimado'],
                title='Cotação Média Mensal: FSoja e DDGS(estimado) 1 ton - média móvel = 3 meses', height=600, width=1200,
                 trendline= "rolling" ,trendline_options=dict(window=3),trendline_scope='trace',
                opacity=1, color_discrete_sequence=px.colors.qualitative.Vivid, template="simple_white",
                 labels={'variable':'cotação', 'value':'BRL/ton', 'Data_FSoja_MT':'dia/mês/ano'})
fig.update_layout(title_font_color="Black", font_size=(14),title_x=0.1)
fig.show()
```
